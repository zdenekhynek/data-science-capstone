```python
import pandas as pd
from gensim import models
import pyLDAvis
import pyLDAvis.gensim

from topics import lda

# get rid of nan
def filter_out_nan(tokens):
    return [token for token in tokens if str(token) != 'nan']

def prepare_lda_vis(tokens_file, lda_model_file):
    # df = pd.DataFrame.from_csv(tokens_file)
    df = pd.read_csv(tokens_file)

    
    tokens = df.values.tolist()
    tokens = [filter_out_nan(row) for row in tokens]
    
    # create dictionary
    dictionary = lda.create_dictionary(tokens)
    dictionary.filter_extremes(1, 0.8)

    # create corpus
    corpus = lda.create_corpus(dictionary, tokens)
    lda_model = models.LdaModel.load(lda_model_file)  
    
    return (lda_model, corpus, dictionary)
    
```

```python
(lda_model, corpus, dictionary) = prepare_lda_vis('data/results/2017-06-27-13-33-40-170445-lda-tokens.csv', 'data/results/2017-06-27-14-11-00-436441-lda.csv')
    
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda_model, corpus, dictionary, mds='mmds')        
    
```
