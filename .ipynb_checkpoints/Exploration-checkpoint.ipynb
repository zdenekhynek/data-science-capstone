```python
# connect to mongo collection
client = MongoClient()
db = client['capstone']
articles_collection = db.articles
```

```python
# create dataframe from articles collection
articles = pd.DataFrame(list(articles_collection.find()))
articles.head()
```

```python
# add column with datetime format converted from timestamp
articles['date'] = pd.to_datetime(articles['webPublicationDate'])
```

```python
# group by year-month
articles['year-month'] = articles.date.dt.year.astype(str) + '-' + articles.date.dt.month.astype(str)
grouped = articles.groupby(['year-month'])
sorted_values = grouped.count().sort_values('_id', ascending=False)
sorted_values
```
