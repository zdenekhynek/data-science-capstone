```python
import pandas as pd

df = pd.DataFrame.from_csv('data/results/2017-06-27-09-09-30-844474-lda-tokens.csv')
tokens = df.values.tolist()

# get rid of nan
def filter_out_nan(tokens):
    return [token for token in tokens if str(token) != 'nan']

tokens = [filter_out_nan(row) for row in tokens]
```

```python
from gensim import models

from topics import lda

dictionary = lda.create_dictionary(tokens)
dictionary.filter_extremes(1, 0.8)
corpus = lda.create_corpus(dictionary, tokens)

# lda_model = models.LdaModel.load('data/results/2017-06-27-07-44-07-295054-lda.csv')
lda_model = models.LdaModel(corpus,
                            num_topics=20,
                            id2word=dictionary,
                            update_every=5,
                            chunksize=100000,
                            passes=100)
```

```python
import pyLDAvis
import pyLDAvis.gensim

pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda_model, corpus, dictionary, mds='mmds')
```
